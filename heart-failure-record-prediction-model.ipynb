```python
X = df.drop("DEATH_EVENT", axis=1)
y = df["DEATH_EVENT"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.svm import SVC
svc = SVC()
svc.fit(X_train, y_train)
y_pred3 = svc.predict(X_test)
accuracy_score(y_test, y_pred3)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
y_pred4 = dtc.predict(X_test)
accuracy_score(y_test, y_pred4)
```

```python
from sklearn.ensemble import RandomForestClassifier
rfr = RandomForestClassifier()
rfr.fit(X_train, y_train)
y_pred5 = rfr.predict(X_test)
accuracy_score(y_test, y_pred5)
```

```python
from sklearn.ensemble import AdaBoostClassifier
ada = AdaBoostClassifier()
ada.fit(X_train, y_train)
y_pred6 = ada.predict(X_test)
accuracy_score(y_test, y_pred6)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbr = GradientBoostingClassifier()
gbr.fit(X_train, y_train)
y_pred7 = gbr.predict(X_test)
accuracy_score(y_test, y_pred7)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knr = KNeighborsClassifier()
knr.fit(X_train, y_train)
y_pred8 = knr.predict(X_test)
accuracy_score(y_test, y_pred8)
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train, y_train)
y_pred9 = xgb.predict(X_test)
accuracy_score(y_test, y_pred9)
```

```python
def heart_failure_record_prediction_model(model, age, anaemia, creatinine_phosphokinase, diabetes,
       ejection_fraction, high_blood_pressure, platelets, serum_creatinine, serum_sodium, sex, smoking, time):
    
            
    x = np.zeros(len(X.columns))
    
    x[0] = age
    x[1] = anaemia
    x[2] = creatinine_phosphokinase
    x[3] = diabetes
    x[4] = ejection_fraction
    x[5] = high_blood_pressure
    x[6] = platelets
    x[7] = serum_creatinine
    x[8] = serum_sodium
    x[9] = sex
    x[10] = smoking
    x[5] = time
    
    
    x = sc.transform([x])[0]
    return model.predict([x])[0]
```
